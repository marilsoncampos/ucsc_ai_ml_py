#### Calculating the area of function

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
import random 
random.seed(100)


class CircleQuadrantSimulator:
    def __init__(self, num_of_experiments):
        self.num_of_experiments = num_of_experiments
        self.x_values = None
        self.y_values = None
        self.wins = None
        self.reset()

    def reset(self):
        self.x_values = []
        self.y_values = []
        self.wins = 0.0
        
    def _pick_a_point(self):
        x = random.random()
        y = random.random()
        return x, y

    def _is_inside_cicle(self, x, y):
        return (x**2 + y**2) <= 1
    
    def simulate(self): 
        for _ in range(self.num_of_experiments):
            x, y = self._pick_a_point()
            if self._is_inside_cicle(x, y):
                self.wins += 1
                self.x_values.append(x)
                self.y_values.append(y)
    
    @property
    def area(self):
        return (self.wins / self.num_of_experiments)


def draw_results(simulator):
    xlin = np.linspace(0,1)
    ylin = []
    for x in xlin:
        ylin.append(math.sqrt(1.0-x**2))
    plt.scatter(simulator.x_values, simulator.y_values, color='lightblue', marker='.', alpha=0.01)
    plt.plot(xlin, ylin, color='red', linewidth='1')

    
sim = CircleQuadrantSimulator(1_000_000)
sim.simulate()
draw_results(sim)
```

```python
print(f'The area of one quadrant = {sim.area:.8f}')
```

Area of the circle = $4 * \pi * radius $

```python
print(f'Estimated pi  = {sim.area * 4:.8f}')
print(f'Calculated pi = {math.pi:.8f}')
```
