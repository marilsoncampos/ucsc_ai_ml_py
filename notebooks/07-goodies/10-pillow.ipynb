```python
from PIL import Image
im = Image.open('sample_image.png')
 
```

```python
print(f'Format = {im.format}')
print(f'Size = {im.size}')
print(f'Mode = {im.mode}')
```

```python
display(im)
```

```python
x1 = 1000
y1 = 50
box = (x1, y1, x1+350, y1+350)
img_box = im.crop(box)
display(img_box)
```

```python
rotated_im = img_box.rotate(1)
display(rotated_im)
```

```python
resized = rotated_im.resize((200, 200))
display(resized)
```
