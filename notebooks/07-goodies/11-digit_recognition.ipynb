#### ML Model to recognize digits.

```python
from sklearn import datasets
from sklearn import svm
import matplotlib.pyplot as plt

def print_digit(idx):
    plt.gray()
    plt.matshow(digits.images[idx])
    plt.show()
```

```python
digits = datasets.load_digits()
```

```python
# Digits represented as matrix of shade values.
digits.data[:]
```

```python
# Answers are just the integer values.
digits.target[:]
```

```python
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(digits.data[:], digits.target[:])
```

```python
predictions = clf.predict(digits.data)
```

```python
NUMBERS_TO_VERIFY = 10
for idx in range(-NUMBERS_TO_VERIFY, 0):
    print('Prediction: {0}'.format(predictions[idx]))
    print_digit(idx)
```
