#### Scipy code for z-score

Given an array of values, scipy.stats(array) will compute the average and standard deviation of the values in the array. It will return the corresponding z-score for each value in the array.

```python
# Scipy example for z-score for elements of numpy array
values = [0.7972,  0.0767,  0.4383,  0.7866,  0.8091, 
          0.1954,  0.6307,  0.6599,  0.1065,  0.0508]
arr_values = np.array(values)
print('Original values:')
print(values)
print('\nZ-scores values:')
print(np.around(stats.zscore(arr_values), 3))
```

```python
# Calculate zcores of a single value using an array representing the distribution. 

def z_score(arr_values, value):
    standard_deviation = np.std(values)
    mean = np.mean(values)
    return (value - mean) / standard_deviation

print(z_score(values, 0.5))
```

### 2. Bernoulli Distribution

If for a random variable X, there are only two outcomes: success and failure

> The Bernoulli distribution of a X is given by $Ber(p)$ 
> where $p$ is the probability of success, 
> and $1-p.$ the probability of failure.

The mean or expected value of Bernoulli distribution is

> $E(X) = p$



#### Variance of Bernouly distribution
The general formula for variance is:

$ var(X) = \sum \limits _{n=1} ^{N} (mean-x_{n})^2 * P(x_{n}) $

N represents the possible values.

For success the term is $ = ((p - 1)^2 p)  $

For failure the term is $ = ((p - 0)^2 (1 - p)) $

Putting the sumation together gives:

$ = ((p - 1)^2 p) + ((p - 0)^2 (1 - p)) $

$ = (p^3 -2p^2 + p) + (p^2 - p^3 ) $

$ = p - p^2 $

Variance is



$ var(X) = p(1 - p) $


```python
plot(lambda p: p * (1 -p), 'p', 'var(X)', 'Variance of Bernoulli Distribution')
```

### 3 Binomial distribution

The binomial distribution describes the probabilities when the Bernoulli trials are repeated.

Reference: https://my.eng.utah.edu/~cs3130/lectures/L07-BernoulliBinomialGeometric.pdf

#### N Choose K calculation

```python
import math

# Few helper functions.

def n_choose_k(n, k):
    numerator = math.factorial(n)
    denominator = math.factorial(n-k) * math.factorial(k)
    return numerator / denominator

def print_result(title: str, value: float) -> None:
    print('{0} = {1}%'.format(title, round(value * 100, 2)))
    
# Verify that nCk = nC(n-k)
n = 7
k = 3
print(n_choose_k(n, k))
print(n_choose_k(n, n-k))
```

**Example 1**

A basketball player traditionally makes 85% of her free throws. Suppose she shoots 10 baskets and counts the number she makes. What is the probability that she makes 8 baskets?


$p = 0.85$

$q = 1-0.85 = 0.15$, 

$n=10$ and $k=8$.

The probability 8 successes out of 10 = 

$10C8 * (0.85)^8 * (0.15)^2$ = 

= $\frac{10!}{(10-8)! * 8!} * (0.85)^8*(0.15)^2 = 0.2759 =$

= $27.59\%$

```python
# Calculation Example 1
ex_1_answer = n_choose_k(10, 8) * math.pow(0.85, 8) * math.pow(0.15, 2)
print_result('Answer example 1', ex_1_answer)
```

**Example 2**

Traditionally, about 70% of students in a particular Statistics course at ECC are successful. 

Suppose 20 students are selected at random from all previous students in this course. 

What is the probability that more than 15 of them will be successful in the course?

$p = 0.7$

$q = 1-0.7 = 0.3$, 

$n=20$ and $k=15$.

The probability 15 successes out of 20 = 

$20C15 * (0.7)^{15} * (0.3)^5$ = 

= $\frac{20!}{(20-15)! * 15!} * (0.7)^{15} * (0.3)^5 = $

= $17.89\%$  

```python
# Calculation Example 2
ex_2_answer = n_choose_k(20, 5) * math.pow(0.7, 15) * math.pow(0.3, 5)
print_result('Answer example 2', ex_2_answer)
```

#### Scipy code for binomial distribution

Using stats.binom.stats, we can compute the mean and variance of a binomial distribution.


```python
# Scipy code for binomial distribution

n = 10
p = 0.4
# Moments values = Mean(m), variance(v), skew(s), and/or kurtosis(k).
moments='mv'
mean, var = stats.binom.stats(n, p, moments=moments)
print("mean: ", mean)
print("variance: ", var)
```

**Example 4**

A factory produces printers of which 2% are defective. Each box contains 10 printers. 
A box is selected at random. 

    - (a) What is the probability that the box contains exactly one defective component?
    - (b) What is the probability that there are at least 2 defective printers in the box?


**Example 4.a**

Given that:

$p=0.02$ and $q=1-0.02=0.98$

$n=10$ and $k=1$

We have: 

$ \text{P for exactly one defective} = 10C1*(0.02)^1*(0.98)^9= 0.1667 = $ 

$ \text{P for exactly one defective} = 16.67\% $

```python
# Calculation for the example 4.a.
ex_4a_answer = n_choose_k(10, 1) * math.pow(0.02, 1) * math.pow(0.98, 9)
print_result('Answer example 4a', ex_4a_answer)
```

**Example 4.b**

Given that:

$p=0.02$ and $q=1-0.02=0.98$

$n=10$ and $k=1$

Also we know from 4.a that: __P(one defective) = 0.1667__

Notice that __P(more two or more defective) = 1 - P(one defective) - P(no defective)__

Then we calculate  P(no defective)

$ \text{P no defective} = 10C0 * (0.02)^0 * (0.98)^{10} = 0.8171 $

$ \text{P of at least two defective} = 1- \text{P no defective} - \text{P one defective} = 1 - 0.8171 - 0.1668=1.61\% $

```python
# Calculation for the P(no defective)
p_no_defective = n_choose_k(10, 0) * math.pow(0.02, 0) * math.pow(0.98, 10)
print_result('P(no defective)', p_no_defective)
print_result('P(At least two defective)', 1 - 0.8171 - 0.1668)
```

```python
# General function to calculate binomial probabilities
def bin_solution(p, n, k):
    res = n_choose_k(n,k) * pow(p, k) * pow(1-p, n-k)
    return round(res, 4)

# 101  (0.02)^1 (0.98)^9
print_result('P(One deffective)', bin_solution(0.02, 10, 1))
# 100  (0.02)^0 (0.98)^10 = 1 * 1 * (0.98)^10
print_result('P(No deffective)', bin_solution(0.02, 10, 0))
```

### 4 Geometric distribution

The geometric distribution involves repeating Bernoulli trials $k$ times, where failures occurs for the first $k-1$ trials and the first success occurs on the last trial, which is the $k^{th}$ trial.

Mean of geometric distribution is 

$E(y) = \mu = \frac{1}{p}$

and the variance of y is

$ var(y) = \sigma^2 = \frac{1-p}{p^2}$

```python
plot(lambda p: (1.0-p)/(p*p), 'p', 'var(X)', 'Variance of Geometric Distribution', x_start=0.1)
```

### 5. Uniform Distribution

A uniform distribution is a distribution that has a constant probability. 

For example:  If we roll a die(numbered from 1 to 6), then the probability of getting 1 is one out of six, i.e., 1/6

Similarly, the probability of getting 2, 3, 4, 5, and 6 also is . There is an equal chance of getting each of the six outcomes.

Now, if we check for the probability of getting 7, it is 0 since it is impossible to get a 7 when rolling a die.

We have an equal chance of occurrence for the probability of outcomes from 1 to 6, and this is what we call a Discrete Uniform Distribution.

Remember that the sum of their probabilities equals 1 or 100%.
