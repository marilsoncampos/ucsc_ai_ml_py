## Import packages

```python
import pandas as pd
import numpy as np
# Loading the plotting libraries.
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading datafrane

```python
train_df = pd.read_csv('../data/train.csv')
```

## Draw scatter Age x Fare - PClass

```python
sns.scatterplot(data=train_df, x='Age', y='Fare', hue='Pclass')
plt.show()
```

## Draw scatter Age x Fare - Survived

```python
sns.scatterplot(data=train_df, x='Age', y='Fare', hue='Survived')
plt.show()
```

## Draw graphs from function code

```python
grid = sns.FacetGrid(train_df, col='Survived', row='Pclass', height=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', color="salmon", alpha=.5, bins=20)
grid.add_legend()
plt.show()
```

```python
sns.relplot(x='Fare', y='Age', hue='Survived', style='Survived',
            col='Sex', data=train_df, height=4)
plt.show()
```
