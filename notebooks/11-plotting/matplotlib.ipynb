```python
import numpy as np
import matplotlib.pyplot as plt


def print_header(title='', sep='.', witdh=46):
    '''Prints a formated header.'''
    title_size = len(title)
    filler_size = (witdh - (title_size + 2)) // 2
    filler_str = sep * filler_size
    if len(title) == 0:
        rfill = filler_str
    else: 
        rfill = filler_str[::-1]
        if (title_size % 2 == 1):
             rfill = rfill + rfill[0]
    print('{left}[{title}]{right}'.format(
        left=filler_str, title=title, right=rfill))

def print_separator():
    print('. ' * 25)
```

Using two arrays of values (t1 and t2)

- We are defining subplots. 
- Subplots are used to show two or more plots in a single image. 


```python
def exp_and_cosine(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
fig = plt.figure("first_plot")

# 211 in subplot() means 2 plots, and 11 is for the location of the plot. 
# since we want to plot two graphs, have to specify the location for each plot.
# The 11 and 12 options indicate that the plots should be stacked vertically. 
plt.subplot(211)
plt.plot(t1, exp_and_cosine(t1), "bo")



# The number 212 in subplot if for the second plot. 12 represents the position of the second plot.
# The x-values of the plot are t2, the y-values are cosine of t2 with slight variation. "r--" 
# guarantees red dashes. 
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), "g--")
fig.savefig('cosine.png')
plt.show()
```

```python
# Create subplots using axes
fig, ax = plt.subplots(2, 2, sharex='col', sharey='row')
ax[0][0].plot(t1, exp_and_cosine(t1), "bo")
ax[0][1].plot(t2, np.cos(2*np.pi*t2), "rx")
ax[1][0].plot(t2, np.cos(2*np.pi*t2), "y--")
ax[1][1].plot(t2, np.cos(2*np.pi*t2), "g.")
```

```python
# 121 and 122 mean two plots rendered horizontally. 

plt.figure("my_graph2", figsize=(18, 3)) 
plt.subplot(121) 
# since we want to plot two graphs, have to specify the location for each plot.
plt.plot(t1, exp_and_cosine(t1), "bo")

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
plt.show()
```

### Adding Text to the figure.

```python
np.random.seed(2021)
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# Setting the histogram
n, bins, patches = plt.hist(x, 50, density=True, facecolor = "r", alpha = 0.2)
# Decorating the plot.
plt.xlabel("Smarts")
plt.ylabel("Probability")
plt.title("Histogram of IQ")
plt.text(60, 0.025, r'$\mu=100, \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### Plot with annotations



Are assiging the plot to the variable called line.

We use annotate() and supply the text that should be added inside the plot. In this case, 
the annotated text is "local max". 
xy = (2,1) specifies the location of the arrow, 
xytext=(3, 1.5) specifies the location of the text. 
A dictionary with one key-value pairs is assigned to arrowprops. 
facecolor is for the color of the arrow. 

in ylim we specify the height of the plot.



```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line = plt.plot(t, s, "r")

plt.annotate("A local minima", xy = (2.5, -1), xytext=(2.8, -1.5), arrowprops = dict(facecolor="green"))
plt.ylim(-2, 2)
plt.show()
```

#### Shading the region under the graph.

- The fill_between() we specify the region to bound, which is x, 0 and y. 
- The value of alpha is used to lighten the color. 


```python
n = 256
x = np.linspace(-np.pi,np.pi,n,endpoint=True)
y = np.cos(2*x)
plt.plot (x, y, color='b', alpha=1.00) # alpha 1.00 means original shade of the color
plt.fill_between(x, 0, y, color='yellow', alpha=.2) # alpha 0.1 means light shade of the color
plt.show()
```

### Multiple graphs in one figure. 

 - x represents the input. 
 - F1, F2 and F3 are three different outputs.

```python
import matplotlib.pyplot as plt
x = np.linspace(-2 * np.pi, 2 * np.pi, 50, endpoint=True)
F1 = 3 * np.sin(x)
F2 = np.sin(2*x)
F3 = 0.3 * np.sin(x)
startx, endx = -2 * np.pi - 0.1, 2*np.pi + 0.1
starty, endy = -3.1, 3.1
plt.plot(x,F1)
plt.plot(x,F2)
plt.plot(x,F3)
plt.show()
```

#### Graph with lines connecting the points and dots at each point 

Here days represents the x-values. The lists celsius_min and celsius_max are two different y-values.
The xlabel, "Day" will be displayed along the x-axis and ylable, "Degrees Celsius" will be 
displayed along the y-axis.

In the plot() we are giving four sets of x and y. 
xmin, xmax, ymin, ymax = 0, 10, 14, 45 speficies the range for x-axis and range for y-axis 
respectively.


```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')

plt.plot(days, celsius_min,
         days, celsius_min, "oy",
         days, celsius_max, 
         days, celsius_max, "*r") 
print("Limits for the axes are:", end=' ')   
axis_list = ['{:.2f}'.format(x) for x in plt.axis()]
print(', '.join(axis_list))
print("Setting the axes to:", end=' ')
xmin, xmax, ymin, ymax = 1, 9, 14, 40
print(xmin, xmax, ymin, ymax)
plt.axis([xmin, xmax, ymin, ymax])
plt.show()
```

#### Four plots with the same data 
- Different scales for the y-axis.

```python
from matplotlib.ticker import NullFormatter

np.random.seed(2021)

y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y<1)]
y.sort()
x = np.arange(len(y))


plt.figure(1, figsize=(8, 8))

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale("linear")
plt.title("linear")
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale("log")
plt.title("log")
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale("symlog")
plt.title("symlog")
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale("logit")
plt.title("logit")
plt.grid(True)

# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1-10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

#### Stacked plots on the same graph.

```python
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
yrs = 1950 + rng
print_header('Years')

print(yrs)
print_header('X random nums')
print(rnd)
print_header('Y randoms nums')
random_y = rng+rnd
print(random_y)

fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])

fig.tight_layout()
```

### Scatter plot and a bar graph.

```python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
print_header('x values')
print(x)
print_header('y values')
print(y)
data = np.column_stack((x, y)) # column_stack does vertical stacking.
print_header('stacked data')
print(data[:10])
print_header('Range')
print(np.arange(data.min(), data.max()))
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5)) # figure size

# scatter plot detials
ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='green')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

# bar graph details
ax2.hist(data, bins=np.arange(data.min(), data.max()), label=('$x$', '$y$'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
```

### Pie chart

```python
companies = ["A", "B", "C", "D", "E"]
market_share = [15, 20, 25, 10, 30]
Explode = [0, 0.2, 0, 0, 0]
plt.pie(market_share, explode = Explode, labels=companies, shadow = True, startangle=45)
plt.axis('equal') # adjustes the size of the pie
plt.legend(title="List of Firms")
plt.show()
```

#### Simple scatter plot

```python
study_time = np.array([145, 170, 165, 184, 175, 159, 180, 172])
score = np.array([80, 92, 89, 90, 95, 75, 96, 85])

plt.xlim(140, 190) # x range 
plt.ylim(50, 100) # y range
plt.scatter(study_time, score) 
plt.title("Study Time Vs Score")
plt.xlabel("Study Time")
plt.ylabel("Score")
plt.show()
```

#### 3D plotting

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d') # creates 3D axis with y-values getting projected onto the third axis
z =[2,3,3,3,5,7,9,11]
ax.scatter3D(study_time, score, z)
ax.set_xlabel("Study Time")
ax.set_ylabel("Score")
plt.show()
```
